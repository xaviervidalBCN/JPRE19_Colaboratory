```
#@title Informació del track { run: "auto" }
track    = 0 
channel  = 0
time     = 0    # In beats
duration = 1    # In beats
Tempo_en_BPM = 200 #@param {type:"slider", min:10, max:1000, step:10} in BMP
Volume = 100 #@param {type:"slider", min:0, max:127, step:1} 0-127, as per the MIDI standard
```

```
#@title Definició de notes 
# 48 = C4 (do central)

Escala = 47 + np.array([17, 1, 3, 5, 6, 8, 10, 12, 13, 15]) #1 = 1, 3= 2, 5= 3 ...
print(Escala)
```

```
#@title Carregar 'generador' amb les notes...

Base  = "3141592653589793238462643383279502884197169399375105820974944592307816406286208998628034825342117067982148086513282306647093844609550582231725359408128481117450284102701938521105559644622948954930381964428810975665933446128475648233786783165271201909145648566923460348610454326648213393607260249141273724587006606315588174881520920962829254091715364367892590360011330530548820466521384146951941511609433057270365759591953092186117381932611793105118548074462379962749567351885752724891227938183011949129833673362440656643086021394946395224737190702179860943702770539217176293176752384674818467669405132000568127145263560827785771342757789609173637178721468440901224953430146549585371050792279689258923542019956112129021960864034418159813629774771309960518707211349999998372978049951059731732816096318595024459455346908302642522308253344685035261931188171010003137838752886587533208381420617177669147303598253490428755468731159562863882353787593751957781857780532171226806613001927876611195909216420198"
Base2 = "1234567890"

generador = Base2
```

```
#@title Codificar el Tema

degrees = np.empty(0, dtype=int)

n = len(generador)
for i in range(0,n):
  decimal = int( generador[i] )
  nota = Escala [decimal]
  degrees = np.append(degrees, nota )
  
print(degrees[:10])
```

```
#@title Creació de l'arxiu MIDI
MyMIDI = MIDIFile(1)  # One track, defaults to format 1 (tempo track is created
                      # automatically)
MyMIDI.addTempo(track, time, Tempo_en_BPM)

for i, pitch in enumerate(degrees):
    MyMIDI.addNote(track, channel, pitch, time + i, duration, Volume)
   
Nom_arxiu = "Tema_ok.mid" #@param {type:"string"}
  
with open(Nom_arxiu, "wb") as output_file:
    MyMIDI.writeFile(output_file)

files.download(Nom_arxiu)
```
